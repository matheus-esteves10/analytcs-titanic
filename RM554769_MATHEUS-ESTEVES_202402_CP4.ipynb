##Qual é a quantidade de embarques de cada porto presente no dataset?

```python
dados = df_titanic['Embarked']
dados_numericos = dados.map({'C': 2,'Q': 1, 'S': 0})

contador_q = 0
contador_c = 0
contador_s = 0

for item in dados_numericos:
  if item == 1:
    contador_q += 1
  elif item == 2:
    contador_c += 1
  elif item == 0:
    contador_s += 1

print(f'Quantidade de embarques no porto Queenstown: {contador_q}')
print(f'Quantidade de embarques no porto Cherbourg: {contador_c}')
print(f'Quantidade de embarques no porto Southampton: {contador_s}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

##Demonstre com um gráfico a quantidade de sobreviventes e não sobreviventes de acordo com classe de embarque.

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df_titanic, x='Pclass', hue='Survived')

plt.legend(title='Sobreviveu', loc='upper left')

plt.title('Distribuição das Classes de Passageiros por Sobrevivência')
plt.xlabel('Classe')
plt.ylabel('Número de Passageiros')
plt.xticks(ticks=[0, 1, 2], labels=['1ª Classe', '2ª Classe', '3ª Classe'])
sns.set_style("whitegrid")
plt.show()
```

Com esse gráfico de barras podemos interpretar que as pessoas que estavam em classes melhores tiveram um privilégio no quesito de sobrevivência, onde na priemira classe a maior parte sobreviveu, na segunda praticamente houve um equilibrio e na terceira temos uma diferença muito grande, com a maioria não sobrevivendo
